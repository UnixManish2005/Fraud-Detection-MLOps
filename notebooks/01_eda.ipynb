# Credit Card Fraud Detection — Exploratory Data Analysis

This notebook walks through:
1. Dataset overview
2. Class imbalance visualisation
3. Feature distributions
4. Correlation heatmap
5. Amount distribution by class

```python
import sys, os
sys.path.insert(0, os.path.join('..'))  # project root

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted')
%matplotlib inline
```

```python
df = pd.read_csv('../data/creditcard.csv')
print(df.shape)
df.head()
```

```python
# Class distribution
counts = df['Class'].value_counts()
labels = ['Normal (0)', 'Fraud (1)']
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].bar(labels, counts.values, color=['steelblue', 'tomato'])
axes[0].set_title('Class Distribution (count)')

axes[1].pie(counts.values, labels=labels, autopct='%1.2f%%',
            colors=['steelblue', 'tomato'], startangle=90)
axes[1].set_title('Class Distribution (%)')
plt.tight_layout()
plt.show()
print(counts)
```

```python
# Amount distribution by class
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
for cls, ax, color in zip([0, 1], axes, ['steelblue', 'tomato']):
    data = df[df['Class'] == cls]['Amount']
    ax.hist(data, bins=60, color=color, edgecolor='white', alpha=0.85)
    ax.set_title(f'Amount Distribution — {"Normal" if cls == 0 else "Fraud"}')
    ax.set_xlabel('Amount')
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap of V features
v_cols = [c for c in df.columns if c.startswith('V')]
corr = df[v_cols + ['Amount', 'Class']].corr()

plt.figure(figsize=(16, 12))
sns.heatmap(corr, cmap='coolwarm', center=0, linewidths=0.3,
            annot=False, square=True)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Top features correlated with Class
corr_with_class = corr['Class'].drop('Class').abs().sort_values(ascending=False)
corr_with_class.head(15).plot(kind='bar', figsize=(12, 4),
                               color='steelblue', edgecolor='white')
plt.title('Top 15 Features Correlated with Fraud (|r|)')
plt.ylabel('|Pearson r|')
plt.tight_layout()
plt.show()
```

```python
# Transaction count over time
plt.figure(figsize=(14, 4))
plt.scatter(df[df['Class']==0]['Time'], df[df['Class']==0]['Amount'],
            alpha=0.05, s=1, color='steelblue', label='Normal')
plt.scatter(df[df['Class']==1]['Time'], df[df['Class']==1]['Amount'],
            alpha=0.5, s=10, color='tomato', label='Fraud')
plt.xlabel('Time (seconds)')
plt.ylabel('Amount')
plt.title('Transactions over Time')
plt.legend()
plt.tight_layout()
plt.show()
```
